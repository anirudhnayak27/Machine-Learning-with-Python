passing '*' in fuction can allow to pass multiple parameters

```python
def args(*arguments):
  sum=0
  for i in arguments:
    sum+=i
  print(sum)

args(2,3,4,5,7)
```

passing '**' in fuction can allow to pass multiple parameters in pairs

```python
def data(**kwargs):
  for key,value in kwargs.items():
    print(key,value)
data(name='John',age=25,gender='male')
```
