```python
df=pd.read_csv("13.1 heart-disease.csv")
df
```

```python
df["age"].plot.hist();
```

```python
df.plot.hist(subplots=True,figsize=(10,40));
```

```python
df_50=df[df['age']>50]
df_50
```

```python
#pyplot method 
df.plot(kind='scatter',x='age',y='chol',c='target');
```

```python
# oo method with pyplot
fig,ax=plt.subplots(figsize=(15,10))
df.plot(kind='scatter',x='age',y='chol',c='target',ax=ax);
ax.set_xlim(20,80);
ax.set_ylim(100,600);
```

```python
##oo method from scratch
fig,ax=plt.subplots(figsize=(15,10))
#ploting data
scatter=ax.scatter(x=df_50['age'],
                   y=df_50['chol'],
                   c=df_50['target']);
#customizing plot
ax.set(title='cholestrol vs age',xlabel='age',ylabel='cholestrol');
ax.legend(*scatter.legend_elements(),title='Targets');
#setting horizontal line
ax.axhline(df_50['chol'].mean(),linestyle='--');
```

```python
## subploting cholage,thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10), sharex=True);
scatter=ax0.scatter(x=df_50['age'],
                   y=df_50['chol'],
                   c=df_50['target']);
ax0.set(title='cholestrol vs age',ylabel='cholestrol');
ax0.legend(*scatter.legend_elements(),title='Targets');
ax0.axhline(df_50['chol'].mean(),linestyle='--');

scatter=ax1.scatter(x=df_50['age'],
                   y=df_50['thalach'],
                   c=df_50['target']);
ax1.set(title='max heart rate vs age',xlabel='age',ylabel='cholestrol');
ax1.legend(*scatter.legend_elements(),title='Targets');
ax1.axhline(df_50['thalach'].mean(),linestyle='--');


fig.suptitle("Heart Diseases Analysis",fontsize=15,fontweight="bold");
```

```python
##oo method from scratch
#selecting style
plt.style.use("seaborn-v0_8-darkgrid")

fig,ax=plt.subplots(figsize=(15,10))
#ploting data
scatter=ax.scatter(x=df_50['age'],
                   y=df_50['chol'],
                   c=df_50['target'],
                  cmap='summer_r');#changes colour scheme
#customizing plot
ax.set(title='cholestrol vs age',xlabel='age',ylabel='cholestrol');
ax.legend(*scatter.legend_elements(),title='Targets');
#setting horizontal line
ax.axhline(df_50['chol'].mean(),linestyle='--');
```

```python
## subploting cholage,thalach  (customizing plots )
plt.style.use('seaborn-v0_8')

fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,15), sharex=True);
#ploting data
scatter=ax0.scatter(x=df_50['age'],
                   y=df_50['chol'],
                   c=df_50['target'],
                   cmap='winter');
#customizing plot
ax0.set(title='cholestrol vs age',ylabel='cholestrol');
ax0.set_xlim([50,80])
ax0.legend(*scatter.legend_elements(),title='Targets');
ax0.axhline(df_50['chol'].mean(),linestyle='--',color='red');

#ploting data
scatter=ax1.scatter(x=df_50['age'],
                   y=df_50['thalach'],
                   c=df_50['target'],cmap='winter');
#customizing plot
ax1.set(title='max heart rate vs age',xlabel='age',ylabel='cholestrol',ylim=(65,200));
ax1.legend(*scatter.legend_elements(),title='Targets');
ax1.axhline(df_50['thalach'].mean(),linestyle='--',color='red');

#adding figure title
fig.suptitle("Heart Diseases Analysis",fontsize=15,fontweight="bold");
```
