Linear Regression using random data generated from 'random module'

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
x=np.random.randint(1,100,10).reshape(-1,1)
y=np.random.randint(1,100,10).reshape(-1,1)
print("x:\n",x)
print("Y:\n",y)
```

```python
plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

importing linear regression fuction from sklearn's linear model.Also fitting the model for our data (x,y)

```python
from sklearn.linear_model import LinearRegression
slr=LinearRegression()
slr.fit(x,y)
```

Predict value of y for input x

```python
predict=slr.predict(x)
predict
```

Find mean square error,slope and intercept .

```python
from sklearn.metrics  import mean_squared_error
mse=mean_squared_error(y,predict)
plt.plot(x,predict,marker='*')
plt.scatter(x,y, color="green",label="Actual without fit")
plt.scatter(x,predict, color="red",label="Predicted with fit")
plt.legend(loc="center right")
print("MSE:",mse)
print("Intersept:",slr.intercept_)
print("Slope:",slr.coef_)
```

Linear regression using data from csv file.

```python
df=pd.read_csv("LR_data.csv")
df
```

```python
from sklearn.linear_model import LinearRegression
slr=LinearRegression()
x=np.array(df["x"]).reshape(-1,1)
y=np.array(df["y"]).reshape(-1,1)
slr.fit(x,y)
```

```python
predict=slr.predict(x)
from sklearn.metrics  import mean_squared_error
mse=mean_squared_error(y,predict)
plt.plot(x,predict,marker='*')
plt.scatter(x,y, color="green",label="Actual without fit")
plt.scatter(x,predict, color="red",label="Predicted with fit")
plt.legend(loc="center right")
print("MSE:",mse)
print("Intersept:",slr.intercept_)
print("Slope:",slr.coef_)
```
