#Rock-Paper-Scissors
User can enter choice betweeen rock,paper or scissors.

user can quit by entering y or continue by entering n.

```python
import numpy as np
import random

def user_choice():
 while True:
  user=input("Enter 'Rock','Paper' or 'Scissor' : ").lower()
  if user in ['rock','paper','scissor']:
    return user
  else :
   print("Invalid choice, choose again. ")

def computer_choice():
  choice=['rock','paper','scissor']
  return random.choice(choice)

def choose_winner(user,comp):
  if user==comp:
    return 'draw'
  elif ((user=='rock' and comp=='scissor') or (user=='paper' and comp=='rock')
         or (user=='scissor' and comp=='paper')):
    return 'player'
  else:
    return 'comptuter'
draw=0
player=0
computer=0
while True:
  user=user_choice()
  print(f"user choice: {user}")
  comp=computer_choice()
  print(f"computer choice: {comp}")
  win=choose_winner(user,comp)
  if win=='draw':
    draw += 1
    print("Its a draw.")
  elif win=='player':
    player += 1
    print("User won.")
  else:
    computer += 1
    print("computer won.")
  quit=input("Want to quit [y/n]: ").lower()
  if quit=='y':
    break
print(f'player wins: {player}')
print(f'computer wins: {computer}')
print(f'draw : {draw}')
```

##Write result in csv file

```python
import pandas as pd
game={'Player Wins':[player],'Computer Wins':[computer],'Draw':[draw]}
df=pd.DataFrame(game)
df.to_csv("Rock-paper-scissor.csv",index=False)
```
